Cleaning AppleStore data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#loading the data
raw = pd.read_csv('AppleStore.csv')
df = raw.copy()

df.shape
df.dtypes
df.info()
df.describe()
df.describe(include = 'object')

plt.scatter(df['user_rating'], df['rating_count_tot'], alpha=0.6)
plt.xlabel('user_rating')
plt.ylabel('rating_count_tot')
plt.title('Rating vs Total Ratings')
plt.show()
```
