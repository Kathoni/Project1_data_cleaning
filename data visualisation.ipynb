Cleaning AppleStore data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#loading the data
raw = pd.read_csv('AppleStore.csv')
df = raw.copy()

df.shape
df.dtypes
df.info()
df.describe()
df.describe(include = 'object')

#checking duplicates 
df.duplicated().sum()
df = df.drop(columns =['Unnamed: 0'])

```

```python
top_genres = df['prime_genre'].value_counts().head(10)
top_genres
```

```python
sns.scatterplot(x='user_rating', y='rating_count_tot', data=df)
plt.title('Rating vs Total Ratings')
plt.show()
```
